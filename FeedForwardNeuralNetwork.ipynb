```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_iris
```

```python
datasets = load_iris()
```

```python
data = datasets.data
target = datasets.target
```

```python
print(data[0:10])
```

```python
target[40:150]
```

```python
print("Number of Unique class : ", np.unique(target))
```

```python
# both have to be equal
print("Length of the Target columns : ", len(target))
print("Length of the data columns :", len(data))
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential() # Empty space
# input layer
model.add(Dense(8, input_dim = 4, activation = 'relu'))
# 1st Hidden layer
model.add(Dense(8, input_dim = 8, activation = 'relu'))
# Final layer
model.add(Dense(3, input_dim = 8, activation = 'softmax'))
```

```python
model.compile(loss = "categorical_crossentropy", optimizer = "sgd", metrics = ["accuracy"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.2)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
categorical_target = to_categorical(y_train)
```

```python
# print(categorical_target)
```

```python
# bach size = epochs 
history = model.fit(x_train, categorical_target, epochs = 200)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['accuracy'])
plt.show()
```

```python
predicted_target = model.predict(x_test)
```

```python
print(predicted_target)
```

```python
print("Pridected :", np.argmax(predicted_target, axis = 1))
print()
print("Actual:    ", y_test)
```
